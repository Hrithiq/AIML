```python
import pandas as pd
import numpy as np
from textblob import TextBlob
```

```python
df=pd.read_csv('flipkart_product.csv',encoding="unicode_escape")
```

```python
df.dropna(inplace=True)
```

```python
df=df.drop(df.loc[df.Rate.isin(['Pigeon Favourite Electric Kettle?ï¿½?ï¿½(1.5 L, Silver, Black)',
       'Bajaj DX 2 L/W Dry Iron',
       'Nova Plus Amaze NI 10 1100 W Dry Iron?ï¿½?ï¿½(Grey & Turquoise)',
       's'])].index)
df.Rate=df.Rate.astype(int)
```

```python
df.Review=df.Review.str.lower()
```

```python
import re
def remove_chars(x):
    s = re.sub('[^0-9]', '', x)
    return s
df.Price = df.Price.apply(remove_chars).astype(int)
```

```python
X=df.drop(['Review','Summary'],axis=1)
```

```python
X=X.groupby('ProductName').mean()
```

```python
X
```

```python
def sentiment(x):
    if x>0.05:
        return 'Positive'
    elif x<0.05 and x>-0.05:
        return 'Neutral'
    else:
        return 'Negative'
```

```python
X['ploarity(Review)']=X['ploarity(Review)'].apply(sentiment)
X['ploarity(Summary)']=X['ploarity(Summary)'].apply(sentiment)
```

```python
X
```

```python
import sklearn
from sklearn.feature_extraction.text import TfidfVectorizer 
from sklearn.metrics.pairwise import cosine_similarity

def product_similarity(x,X):
    l=[]
    name=[]
    vectorizer=TfidfVectorizer()
    for i in X:
        vectors=vectorizer.fit_transform([str(x),i])

        l.append(cosine_similarity(vectors)[0][1])
        name.append(i)

    return pd.DataFrame({'Similarity':l},index=name)
sim=product_similarity(X.index[20],X.index)
```
