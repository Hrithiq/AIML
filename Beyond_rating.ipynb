```python
movies_df=pd.read_csv('movie.csv')
movies_df
```

```python
movies_df.set_index('movieId',inplace=True)
```

```python
movie=movies_df.drop(['title'],axis=1)
```

```python
movie.genres=movie.genres.str.lower().str.split('|')
```

```python
movie.head()
```

```python
from sklearn.preprocessing import MultiLabelBinarizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
classes=movie.genres.explode().unique()

mlb=MultiLabelBinarizer()
```

```python
encod=pd.DataFrame(mlb.fit_transform(movie.genres),columns=mlb.classes_,index=movie.index)
X_enc=pd.concat([movie,encod],axis=1).drop(['genres'],axis=1).astype(float)
```

```python
encod
```

```python
X_enc
```

```python
def recommend(x,X,n):
    sim=cosine_similarity(x,X).reshape(-1,)
    sim=pd.Series(sim,index=X.index,name='Sim')
    X_new=pd.concat([X,sim],axis=1)
    sort=X_new.sort_values(by='Sim',ascending=False)
    return sort.Sim.head(n)
```

```python
movies_df.iloc[3]
```

```python
sim_df=recommend(X_enc.iloc[3,:].values.reshape(1,-1),X_enc,10)
sim_df
```

```python
pd.concat([movie.loc[movie.index.isin(list(sim_df.index))],sim_df],axis=1).sort_values(by='Sim',ascending=False)
```
